## Approach

- Starting by mapping all the values in the __quality__ feature to 0 and 1 based on the condition given. Next, I have normalized all the values in the other columns except Id since it seems to be a feature for identification.

```python
import pandas as pd
import numpy as np

df= pd.read_csv('data/WineQT.csv')
df.sample(5)
```

```python
df['quality'].value_counts()
```

```python
def helper(val):
  if(val>6):
    return 1
  else:
    return 0

df['quality']= df['quality'].map(helper)
```

```python
for column in df.columns:
  if column not in ["quality", "Id"]:
    mn= df[column].min(); mx= df[column].max()
    df[column]= (df[column]- mn)/(mx-mn)
```
